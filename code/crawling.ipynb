# 네이버 증권 크롤링

```python
url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1", headers=headers)
```

```python
def make_table(url):  
    html = BeautifulSoup(url.text)
    table = html.find("table", class_="type_2")
    table = pd.read_html(str(table))[0]

    del table["N"]
    del table["토론실"]

    table = table[table["종목명"].notnull()]
    return table
```

```python
kospi_page = int(html.find("td", class_="pgRR").find('a')['href'][-2::])
kospi_page
```

```python
from tqdm import tqdm # for문의 진행상황 확인

# 코스피, 코스닥 모두 합쳐서 하나의 데이터
kospi_total = []
kosdaq_total = []

for i in range(2):
    if i==0:
        print("-----------코스피 크롤링 시작----------")
        for n in tqdm(range(kospi_page+1)):
            url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.naver?sosok={i}&page{n}1", headers=headers)
            table = make_table(url)
            kospi_total.append(table)
            time.sleep(rd.random() + 1)
    else:
        print("-----------코스닥 크롤링 시작----------")
        for n in tqdm(range(kospi_page+1)):
            url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.naver?sosok={i}&page{n}1", headers=headers)
            table = make_table(url)
            kosdaq_total.append(table)
            time.sleep(rd.random() + 1)
                   
kospi = pd.concat(kospi_total, ignore_index=True)
kospi["소속"] = "KOSPI"
kosdaq = pd.concat(kosdaq_total, ignore_index=True)
kosdaq["소속"] = "KOSDAQ"

stock = pd.concat([kospi, kosdaq], ignore_index=True)
kospi.to_excel("koreaStock.xlsx")
print(f'완료.')
stock
```

```python
url = requests.get("https://finance.naver.com/sise/dividend_list.naver?&page=1", headers=headers)
html = BeautifulSoup(url.text, 'html.parser')
last_page = int(html.find("td", class_="pgRR").find('a')["href"][-2::])
```

```python
total = []

for n in range(last_page):
    url = requests.get("https://finance.naver.com/sise/dividend_list.naver?&page={n}", headers=headers)
    html = BeautifulSoup(url, 'html.parser')


    table = html.find("table", class_="type_1 tb_ty")
    table = pd.read_html(str(table), header=1)[0]

    table = table[table[["종목명", "1년전", "2년전", "3년전"]].notnull()]
    
    pd.to_numeric(table[["1년전", "2년전", "3년전"]], error="corce")
    table = table[(table["1년전"]>table["2년전"]) & (table["2년전"]>table["3년전"])]

    total.append(table)
    table = pd.concat(total, ingore_index=True)

table
```

```python
table = table[(table["1년전"]>table["2년전"]) & (table["2년전"]>table["3년전"])]
table

t = ''
for i in range(len(table)):
    a = table.iloc[i]
    t += (f'{a["종목명"]} : {a["3년전"]}원) → {["종목명"]} : {a["2년전"]}원 → {a["종목명"]} : {a["1년전"]}원 \n')
    f = open("배당금 3년 연속 상승 기업.txt", 'w')
    f.write(t)
    f.close()
```
