```python
# 지도 학습: 문제지, 정답지

import numpy as np

box = np.array([1, 2, 3, 4, 5]) # numpy를 사용해 병렬 연산 (그냥 list는 병렬연산 불가능)
print(box + 5)
print(box * 5)
```

```python
# 가설: 공부시간에 성적에 영향을 미친다.

x1 = np.array([2, 4, 6, 8]) # 공부 시간
x2 = np.array([1, 2, 1, 3]) # 문제집 푼 개수
y = np.array([20, 55, 45, 80])  # 성적

a1 = 0
a2 = 0
b = 0
lr = 0.005

for n in range(3000):
    yhat = a1*x1 + a2*x2 +b
    error = yhat - y
    
    a1_diff = sum(2 * error * x1)
    a2_diff = sum(2 * error * x2)
    b_diff = sum(2 * error * 1)

    a1 = a1 - a1_diff * lr
    a2 = a2 - a2_diff * lr
    b = b - b_diff * lr
    print(f'{n+1}회 학습 // a1: {a1_diff}, a2: {a2_diff}, b: {b_diff}')
```

```python
import matplotlib.pyplot as plt

plt.scatter(x1, y)
plt.plot(x1, a1*x1+a2*x2+b, 'red')
```

```python
a1*5 + a2*2 + b
```

```python
#  사인 그리기

x = np.arange(-3, 3, 0.01)
y = np.sin(x)

a = 0
b = 0
c = 0
d = 0
lr = 0.00001

for n in range(5000):
    yhat = a*x**3 + b*x**2 + c*x + d
    error = yhat - y
    
    a_diff = sum(2 * error * x**3)
    b_diff = sum(2 * error * x**2)
    c_diff = sum(2 * error * x)
    d_diff = sum(2 * error * 1)

    a -= a_diff * lr
    b -= b_diff * lr
    c -= c_diff * lr
    d -= d_diff * lr
    print(f'{n+1}번째 학습// a: {a_diff:.2f}, b: {b_diff:.2f}, c: {c_diff:.2f}, d: {d_diff:.2f}')
```

```python
plt.plot(x, y)
plt.plot(x, a*x**3 + b*x**2 +c*x + d, 'red')
```
